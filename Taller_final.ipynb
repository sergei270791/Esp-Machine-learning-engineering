```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Boxplots para detectar outliers
for col in numeric_cols:
    plt.figure()
    plt.boxplot(df[col].dropna())
    plt.title(f"Boxplot de {col}")
    plt.ylabel(col)
    plt.show()

# Calcular l√≠mites IQR y mostrar valores fuera de rango por variable
outlier_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    outlier_summary[col] = {
        "lower_bound": lower,
        "upper_bound": upper,
        "n_outliers": len(outliers),
        "outlier_samples": outliers.head().tolist()
    }

outlier_summary
```

```python
# Feature Engineering

import pandas as pd
import numpy as np

# 1. Ratio reclamos por producto
df["reclamos_por_producto"] = df["reclamos_ult_6m"] / df["num_productos"]

# 2. Segmento de edad
df["segmento_edad"] = pd.cut(
    df["edad"],
    bins=[17, 30, 45, 60, 100],
    labels=["Joven", "Adulto", "Maduro", "Senior"]
)

# 3. Segmentaci√≥n de ingreso (bajo / medio / alto)
df["segmento_ingreso"] = pd.qcut(df["ingreso_mensual"], q=3, labels=["Bajo", "Medio", "Alto"])

# 4. Cliente digital full
df["cliente_digital_full"] = np.where((df["usa_app"] == 1) & (df["usa_web"] == 1), 1, 0)

# 5. Cliente no digital
df["cliente_no_digital"] = np.where((df["usa_app"] == 0) & (df["usa_web"] == 0), 1, 0)

# 6. Interacci√≥n producto x tasa
df["producto_x_tasa"] = df["num_productos"] * df["tasa_credito"]
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# --- Apply Box-Cox and WOE directly to df (overwrite or add columns) ---

# Ensure respondio_oferta is int for WOE
df["respondio_oferta"] = df["respondio_oferta"].astype(int)

# Numeric columns for Box-Cox
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

for col in numeric_cols:
    # skip constant cols
    if df[col].nunique() <= 1:
        continue
    
    # ensure positive
    col_min = df[col].min()
    if col_min <= 0:
        df[col] = df[col] + abs(col_min) + 1e-6
    
    # apply Box-Cox -> overwrite or create new?
    # We append new column *_boxcox
    df[col + "_boxcox"], _ = stats.boxcox(df[col])

# Categorical columns for WOE
categorical_cols = df.select_dtypes(include=["category", "object"]).columns.tolist()

for col in categorical_cols:
    if col == "cliente_id":
        continue
    
    tmp = df[[col, "respondio_oferta"]].copy()
    grouped = tmp.groupby(col)["respondio_oferta"].agg(["sum", "count"])
    grouped["non_events"] = grouped["count"] - grouped["sum"]
    
    grouped["dist_events"] = grouped["sum"] / grouped["sum"].sum()
    grouped["dist_non_events"] = grouped["non_events"] / grouped["non_events"].sum()
    grouped["woe"] = np.log(grouped["dist_events"] / grouped["dist_non_events"]).replace([np.inf, -np.inf], 0)
    
    # map WOE values into same df
    df[col + "_WOE"] = df[col].map(grouped["woe"].to_dict())
```

```python
# Crear codificaci√≥n ordinal expl√≠cita sobre df para las variables ordinales
import pandas as pd

# Variables ordinales identificadas previamente
ordinal_vars = {
    "nivel_educacion": ["secundaria", "tecnico", "universitario", "posgrado"],
    "satisfaccion": [1, 2, 3, 4, 5]
}

# Crear columnas ordinales codificadas manualmente respetando el orden
for col, order in ordinal_vars.items():
    # Crear un diccionario de mapeo
    ordinal_map = {cat: i+1 for i, cat in enumerate(order)}
    # Crear nueva columna con _ORD
    df[col + "_ORD"] = df[col].map(ordinal_map)

df.head()
```

```python
import matplotlib.colors as mcolors
def agrupaciones_variable(df, modelo, variable, grupos, g1=None, g2=None, g3=None, g4=None):
    """
    Agrupa una variable categ√≥rica de un DataFrame en varios grupos, calculando la efectividad 
    y aplicando colores basados en dicha efectividad.

    Par√°metros:
    - df: DataFrame con los datos originales.
    - modelo: Nombre de la columna target - dependiendo modelo a analizar.
    - variable: Nombre de la variable categ√≥rica que se va a agrupar.
    - grupos: N√∫mero de grupos en los que se dividir√° la variable (2, 3 o 4).
    - g1, g2, g3, g4: Tama√±o espec√≠fico para cada grupo (opcional). Si no se proporciona, los grupos se crean de manera equitativa.

    Retorna:
    - grouped_df: DataFrame con la variable agrupada, efectividad calculada, asignaci√≥n de grupos y colores aplicados en funci√≥n de la efectividad.
    """

    # Agrupar por la variable y calcular el total de ventas y la efectividad
    grouped_df = df.groupby(variable).agg(
        total_registros=(modelo, 'size'),
        target_exitosas=(modelo, 'sum')
    ).reset_index()

    # Calcular el % de efectividad
    grouped_df['%target'] = round((grouped_df['target_exitosas'] / grouped_df['total_registros'] * 100), 1).fillna(0)
    grouped_df = grouped_df.sort_values(by='%target', ascending=False).reset_index(drop=True)
    # % de registros por categor√≠a y % acumulado (global)
    total = grouped_df['total_registros'].sum()
    grouped_df['%registros'] = (grouped_df['total_registros'] / total * 100).round(2)
    grouped_df['%registros_acum'] = grouped_df['%registros'].cumsum().round(2)


    # Si no se proporcionan n√∫meros espec√≠ficos para cada grupo, distribuir equitativamente
    if g1 is None:
        group_sizes = [len(grouped_df) // grupos + (1 if i < len(grouped_df) % grupos else 0) for i in range(grupos)]
    else:
        if grupos == 2 and g2 is None:
            g2 = len(grouped_df)-g1
        elif grupos == 3 and g3 is None:
            g3 = len(grouped_df)-g2
        elif grupos == 4 and g4 is None:
            g4 = len(grouped_df)-g3
        group_sizes = [g1, g2, g3, g4][:grupos]

    # Validar que la suma de las filas asignadas no exceda el n√∫mero total de filas
    if sum(group_sizes) != len(grouped_df):
        raise ValueError("La suma de los tama√±os de los grupos no coincide con el total de registros.")

    # Crear los grupos manualmente basados en los tama√±os especificados
    labels = [f'Grupo {i+1}' for i in range(grupos)]
    group_labels = []
    start = 0
    for i, size in enumerate(group_sizes):
        group_labels.extend([labels[i]] * size)
        start += size

    grouped_df['Grupo'] = group_labels

    # Crear un colormap de matplotlib
    cmap = plt.get_cmap('RdYlGn')
    norm = mcolors.Normalize(vmin=grouped_df['%target'].min(), vmax=grouped_df['%target'].max())

    # Aplicar los colores en funci√≥n de la efectividad
    grouped_df['Color'] = grouped_df['%target'].apply(lambda x: mcolors.rgb2hex(cmap(norm(x))))

    return grouped_df

def tabla_agrupaciones_variable(grouped_df):
    # Crear una tabla formateada con colores, sin mostrar la columna de color
    def color_efectividad(val):
        color = grouped_df.loc[grouped_df['%target'] == val, 'Color'].values[0]
        return f'background-color: {color}; color: black;'

    styled_df = grouped_df.drop(columns=['Color']).style.applymap(color_efectividad, subset=['%target']).format({'%target': "{:.2f}%"}).set_properties(
        **{
            'text-align': 'center',       # Centrar el texto
            'font-size': '10pt',          # Reducir el tama√±o de la fuente
            'padding': '1px 2px'          # Reducir el espacio de relleno en las celdas
        }
    )
    display(styled_df)
```

```python
# Aplicar las reglas de agrupamiento solicitadas sobre df para las variables binned

import pandas as pd

# Vamos a crear nuevas variables de agrupamiento basadas en los criterios del usuario

# --- 1) ingreso_mensual_BIN ---
# Grupos:
# >= 32% -> grupo alto
# >= 28% y < 32% -> grupo medio -- puede ser 30
# < 28% -> grupo bajo

# Primero calculamos %target por bin
ingreso_stats = df.groupby("ingreso_mensual_BIN")["respondio_oferta"].mean().reset_index()
ingreso_stats["pct"] = ingreso_stats["respondio_oferta"] * 100

# Creamos diccionario de mapeo
ingreso_stats["grupo_ingreso"] = pd.cut(
    ingreso_stats["pct"],
    bins=[-1, 28, 32, 100],
    labels=["Bajo","Medio","Alto"]
)

map_ingreso = ingreso_stats.set_index("ingreso_mensual_BIN")["grupo_ingreso"]

df["ingreso_mensual_BIN"] = df["ingreso_mensual_BIN"].map(map_ingreso)


# --- 2) tasa_credito_BIN ---
# Grupos:
# >= 34.7 -> grupo 1
# >= 28.3 & < 34.7 -> grupo 2
# 25.5 & 25 -> grupo 3
# √∫ltimo 22.6 -> grupo 4
# Para simplificar, usamos bins aproximados seg√∫n lo observado.
tasa_stats = df.groupby("tasa_credito_BIN")["respondio_oferta"].mean().reset_index()
tasa_stats["pct"] = tasa_stats["respondio_oferta"] * 100

tasa_stats["grupo_tasa"] = pd.cut(
    tasa_stats["pct"],
    bins=[-1, 25.1, 28.3, 34.7, 100],
    labels=["Muy Bajo","Bajo","Medio","Alto"]
)

map_tasa = tasa_stats.set_index("tasa_credito_BIN")["grupo_tasa"]
df["tasa_credito_BIN"] = df["tasa_credito_BIN"].map(map_tasa)


# --- 3) producto_x_tasa_BIN ---
# Grupos:
# >= 36 -> alto
# >= 28 & < 36 -> medio
# < 28 -> bajo
px_stats = df.groupby("producto_x_tasa_BIN")["respondio_oferta"].mean().reset_index()
px_stats["pct"] = px_stats["respondio_oferta"] * 100

px_stats["grupo_prod_x_tasa"] = pd.cut(
    px_stats["pct"],
    bins=[-1, 28, 36, 100],
    labels=["Bajo","Medio","Alto"]
)

map_px = px_stats.set_index("producto_x_tasa_BIN")["grupo_prod_x_tasa"]
df["producto_x_tasa_BIN"] = df["producto_x_tasa_BIN"].map(map_px)


# Mostrar las primeras filas con nuevas columnas
df[["ingreso_mensual_BIN","ingreso_mensual_BIN",
    "tasa_credito_BIN","tasa_credito_BIN",
    "producto_x_tasa_BIN","producto_x_tasa_BIN"]].head()
```
